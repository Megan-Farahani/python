## Reading main graphs from the file

```python
from fileinput import filename
from igraph import *
import igraph as ig
from igraph import Graph
import os
import pandas as pd


graphs = []

filenames = os.listdir("../../networkRepository")


for f in filenames:
    with open(f"../../networkRepository/{f}", "r") as edges_file:
        edges = []
        line = edges_file.readline()
        
        while line != "":
            strings = line.rstrip().split(" ")
            e = ((int(strings[0]), (int(strings[1]))))
            edges.append(list(e))
            line = edges_file.readline()
    
    
    graphs.append(Graph(edges))

```

## A table to compare algorithms in different graphs from network repository

```python
table_alg_comp = pd.DataFrame(index=filenames)

for i, net in enumerate(graphs): 
    damages = []
    for v in net.vs:
        _,_,d = func.fire(net, v, func.the_greedy)
        damages.append(d)
    table_alg_comp.loc[filenames[i], "the_greedy"] = sum(damages)/len(damages)
    
    
    
for i, net in enumerate(graphs): 
    damages = []
    for v in net.vs:
        _,_,ratio = func.fire(net, v, func.the_greedy_optimized)
        damages.append(ratio)
    table_alg_comp.loc[filenames[i], "the_greedy_optimized"] = sum(damages)/len(damages)
    
for i, net in enumerate(graphs): 
    table_alg_comp.loc[filenames[i], "The number of vertices"] = len(net.vs)

    
    
    
display(table_alg_comp)
```

# Create a Cartesian product of K4 and P3

```python
complete_graph = [] 

i=10 # the number of k4 in the graph
j=4  # the number of vertices k4


for count, x in enumerate(range(1, i*j+1, j)):   # number of all vertices 
    for y in range(x, (count+1)*j):              # vertices in a floor
        for z in range(y+1, ((count+1)*j)+1):    # combination of x and y ( add 1 to each number, then continue)
            complete_graph.append([y-1,z-1])

for w in range(1, (i-1)*j+1):                    # number of columns
    complete_graph.append([w-1,w+j-1])
    
g1 = Graph(edges=complete_graph)    
```

## Visualization

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(16, 9))
vNum=40 # the number of vertices
ig.plot(g1, target=ax, vertex_color=["red"]+["blue"]*(vNum-1), 
            vertex_label_size=18, vertex_label_dist=20, vertex_label=['r']+list(range(1,vNum)))
```

## Diameter of graph g1 with the vertices in the path in order

```python
# the greatest distance between any pair of vertices is diameter(the maximum eccentricity of any vertex in the graph).
# If there are many shortest paths with the length of the diameter, this code returns the first one it founds.

g1.get_diameter(directed=False, unconn=False, weights=None)
```

# Create a three cartesian-product-graph connected with a triangle

```python
# The Cartesian product of graphs is sometimes called the box product of graphs.
# Definition: The graph G \Box H is the Cartesian product of graphs G and H with vertex set
# V (G \Box H) = V (G) × V (H), and for which {(x, u), (y, v)} is an edge if x = y and
# {u, v} ∈ E(H), or {x, y} ∈ E(G) and u = v. 

complete_graph = [] 

i=10 # the number of k4 in the graph
j=4  # the number of vertices k4


for count, x in enumerate(range(1, i*j+1, j)):   # number of all vertices 
    for y in range(x, (count+1)*j):              # vertices in a floor
        for z in range(y+1, ((count+1)*j)+1):    # combination of x and y ( add 1 to each number, then continue)
            complete_graph.append([y-1,z-1])

for w in range(1, (i-1)*j+1):                    # number of columns
    complete_graph.append([w-1,w+j-1])
    
    
i=8
j=4 
l1=40
    
for count, x in enumerate(range(1+l1, i*j+1+l1, j)):   # number of all vertices 
    for y in range(x, (count+1)*j+l1):              # vertices in a floor
        for z in range(y+1, ((count+1)*j)+1+l1):    # combination of x and y ( add 1 to each number, then continue)
            complete_graph.append([y-1,z-1])
    
for w in range(1+l1, (i-1)*j+1+l1):                    # number of columns
    complete_graph.append([w-1,w+j-1])

i=12
j=4 
l2=72
    
for count, x in enumerate(range(1+l2, i*j+1+l2, j)):   # number of all vertices 
    for y in range(x, (count+1)*j+l2):              # vertices in a floor
        for z in range(y+1, ((count+1)*j)+1+l2):    # combination of x and y ( add 1 to each number, then continue)
            complete_graph.append([y-1,z-1])
    
for w in range(1+l2, (i-1)*j+1+l2):                    # number of columns
    complete_graph.append([w-1,w+j-1])

#the number of vertices is 120 so far

#adding paths between three graphs 
complete_graph.extend([[59,24],[109,59],[109,24]])
```

```python
g5 = Graph(edges=complete_graph)
```

## Visualization

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(16, 9))
vNum=120 # the number of vertices
ig.plot(g5, target=ax, vertex_color=["red"]+["blue"]*(vNum-1), 
            vertex_label_size=18, vertex_label_dist=20, vertex_label=['r']+list(range(1,vNum)))
```

## Diameter of graph g5 with the vertices in the path in order

```python
# the greatest distance between any pair of vertices is diameter(the maximum eccentricity of any vertex in the graph).
# If there are many shortest paths with the length of the diameter, this code returns the first one it founds.

g5.get_diameter(directed=False, unconn=False, weights=None)
```

# Define "Fire function spread" in a graph 

```python
def find_neighbors(g5, root, num_vert):
    
    totalBurnt = set([root])
    newly_burnt_vertices = set()
    r=set([root])
    
    i=0
    while(len(r) != 0): # as long as root set is not empty, print and update newly burnt vertices
         
        print(f"round={i}, burning vertices = {r}")
        newly_burnt_vertices.clear()
        
        for j in r:
            n= g5.neighbors(j)
            n = set(n).difference(totalBurnt)
            newly_burnt_vertices.update(n)
            totalBurnt.update(newly_burnt_vertices)
        
        r = newly_burnt_vertices.copy()
        i+=1
```

## Call the fire function spread on the graphs g5

```python
find_neighbors(g5,108,120)   # call the function for root = 108  
```

## Damage caused by a chosen root

```python
n= [len(g5.vs)]
print("The number of vertices: ",n)  


totalBurnt, protected, ratio = func.fire(g5,108, func.the_greedy_optimized)  # root is 108 in more complecated graph g5
print(f"Total burnt vertices: {totalBurnt}, Protected vertices: {protected}, and Damage: {ratio}%")
```
